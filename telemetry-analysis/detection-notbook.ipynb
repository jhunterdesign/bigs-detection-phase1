```python
import json
import glob
from datetime import datetime
from pprint import pprint
```

```python
events = []

import os
print("Notebook location:", os.getcwd())
```

```python
import glob

json_files = glob.glob("events/*.json")
print("Found JSON files:", len(json_files))
json_files[:5]
```

```python
import json

evbents = []

for file in json_files:
    with open(file, "r") as f:
        data = json.load(f)
        events.append(data)

print("Loaded events:", len(events))
```

```python
from pprint import pprint

print("First event structure:\n")
pprint(events[0])
```

```python
event_types = set(e["event_type"] for e in events)
print("Event types seen:")
event_types
```

```python
flagged = [e for e in events if e["detection"]["flagged"]]
print("Flagged events:", len(flagged))
```

```python
not_flagged = [e for e in events if not e["detection"]["flagged"]]
print("Not flagged events:", len(not_flagged))
```

```python
sessions = set(e["session_id"] for e in events)
print("Sessions observed:", sessions)
```

```python
from collections import Counter

counter = Counter(e["event_type"] for e in events)
counter
```

```python
for e in not_flagged:
    print(e["event_type"], " | score:", e["detection"]["score"])
```

```python
for e in flagged:
    print(e["event_type"], " | score:", e["detection"]["score"])
```

```python
scores = [e["detection"]["score"] for e in events]
scores
```
